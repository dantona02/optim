```python
%load_ext autoreload
%autoreload 1
%aimport bmc.utils.tex_variation
from bmc.utils.tex_variation import run_variation_parallel
import numpy as np
import torch

adc_time = 5e-3
webhook = False

low = -1e-3
high = 1e-3
n_iso = 100
z_pos = np.linspace(low, high, n_iso)
z_pos = torch.tensor(z_pos)
z_pos = torch.cat((z_pos, torch.tensor([0.0])))  # Füge 0 hinzu
z_pos = torch.sort(z_pos).values

config_path = 'sim_lib/config_1pool.yaml'
```

### No Phase Advance (1ETM, alpha=30°)

```python
seq_on = 'seq_lib/30no_phasead.seq'
seq_off = 'seq_lib/30no_phasead_minus.seq'

print(f'n_iso: {len(z_pos)}')
_, singal, plot_data = run_variation_parallel(seq_on, seq_off, config_path, adc_time, z_pos, webhook, show_plot=False, save_plot=False, store_all=True)
```

```python
%matplotlib widget
import matplotlib.pyplot as plt
print(f'Signal: {singal.item()}')
# fig

fig, ax = plt.subplots(dpi=150)
for index, t_val in enumerate(plot_data['t']):
    ax.plot(t_val, plot_data['signal_plot'][index], 'o', markersize=1)
ax.axhline(0, c='black')
ax.set_ylim(-0.0008, 0.0003)
plt.show()
```

### Phase Advance 180° (1ETM, alpha=30°)

```python
seq_on_phasead = 'seq_lib/30phasead.seq'
seq_off_phasead = 'seq_lib/30phasead_minus.seq'

print(f'n_iso: {len(z_pos)}')
_, singal_phasead, plot_data_phasead = run_variation_parallel(seq_on_phasead, seq_off_phasead, config_path, adc_time, z_pos, webhook, show_plot=False, save_plot=False, store_all=True)
```

```python
%matplotlib widget
import matplotlib.pyplot as plt
print(f'Signal: {singal_phasead.item()}')
# fig

fig, ax = plt.subplots(dpi=150)
for index, t_val in enumerate(plot_data_phasead['t']):
    ax.plot(t_val, plot_data_phasead['signal_plot'][index], 'o', markersize=1)
ax.axhline(0, c='black')
ax.set_ylim(-0.0008, 0.0003)
plt.show()
```

### Evaluation

```python
%matplotlib widget
import matplotlib.pyplot as plt

fig, ax = plt.subplots(dpi=150)

# No Phase Advance (blau)
for index, t_val in enumerate(plot_data['t']):
    ax.plot(t_val, plot_data['signal_plot'][index], 'o', markersize=1, color='blue', label='No Phase Advance' if index == 0 else None)

# Phase Advance 180° (rot)
for index, t_val in enumerate(plot_data_phasead['t']):
    ax.plot(t_val, plot_data_phasead['signal_plot'][index], 'o', markersize=1, color='red', label='Phase Advance 180°' if index == 0 else None)

ax.axhline(0, c='black', linewidth=0.5)
ax.set_ylim(0.0, 0.0002)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal Difference')
ax.legend()
plt.tight_layout()
plt.show()

print(f'Signal (No Phase Advance): {singal.item():.6f}')
print(f'Signal (Phase Advance 180°): {singal_phasead.item():.6f}')
```
